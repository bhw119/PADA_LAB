# Depth

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
from scipy.stats import entropy

# Step 1: Preprocess reviews and prepare data for LDA
# 리뷰 데이터를 LDA에서 사용할 수 있도록 처리
vectorizer = CountVectorizer(max_features=1000, stop_words='english')
data_bow = vectorizer.fit_transform(df['Review_Text'])  # 리뷰 텍스트를 BoW로 변환
data_vocab = vectorizer.get_feature_names_out()

# Step 2: Fit the LDA model
# LDA 모델 생성 및 학습
lda_model = LatentDirichletAllocation(
    n_components=10,  # 토픽 수
    max_iter=20,      # 반복 횟수
    random_state=42,
    learning_method='online'
)
lda_model.fit(data_bow)

# Step 3: Compute topic proportions for each sentence
# 각 문장에 대한 토픽 비율 계산
def get_sentence_topic_proportions(sentence, model, vectorizer):
    sentence_bow = vectorizer.transform([sentence])  # 문장을 BoW로 변환
    topic_proportions = model.transform(sentence_bow)  # 토픽 비율 계산
    return topic_proportions[0]  # 토픽 비율 반환

# Step 4: Aggregate sentence-level topic proportions to review-level topic proportions
# 리뷰의 문장별 토픽 비율을 전체 리뷰 단위로 평균화
def aggregate_to_review_level(review_text, model, vectorizer):
    sentences = review_text.split('.')  # 문장으로 나누기 (기본적인 분할 방식)
    sentence_proportions = []

    for sentence in sentences:
        if sentence.strip():  # 빈 문장은 제외
            sentence_prop = get_sentence_topic_proportions(sentence.strip(), model, vectorizer)
            sentence_proportions.append(sentence_prop)

    # 전체 리뷰에 대한 평균 토픽 비율 계산
    avg_topic_proportions = np.mean(sentence_proportions, axis=0)
    return avg_topic_proportions

# Step 5: Calculate Shannon entropy for each review
# Shannon 엔트로피 계산
def calculate_entropy(topic_proportions):
    return entropy(topic_proportions, base=2)  # Shannon 엔트로피 계산

# Step 6: Compute ContentDepth for each review
# ContentDepth 계산
def compute_content_depth(review_text, model, vectorizer):
    # 리뷰 단위의 토픽 비율 계산
    avg_topic_proportions = aggregate_to_review_level(review_text, model, vectorizer)
    # 리뷰의 엔트로피 계산
    entropy_value = calculate_entropy(avg_topic_proportions)

    # 문장 수로 엔트로피를 정규화
    num_sentences = len([s for s in review_text.split('.') if s.strip()])

    # ContentDepth는 정규화된 엔트로피
    content_depth = entropy_value / num_sentences if num_sentences > 0 else 0
    return content_depth

# Step 7: Apply ContentDepth calculation to each review in the DataFrame
# DataFrame에 ContentDepth 적용
df['ContentDepth'] = df['Review_Text'].apply(lambda x: compute_content_depth(x, lda_model, vectorizer))

# 결과 보기
df
```

# Breadth

```python
# Step 1: Calculate the average topic proportions across all reviews
# 전체 리뷰의 평균 토픽 비율 계산
def calculate_average_topic_proportions(df, model, vectorizer):
    all_review_proportions = []

    for review in df['Review_Text']:
        avg_topic_proportions = aggregate_to_review_level(review, model, vectorizer)
        all_review_proportions.append(avg_topic_proportions)

    # 모든 리뷰의 평균 토픽 비율 계산
    global_topic_proportions = np.mean(all_review_proportions, axis=0)
    return global_topic_proportions

# Step 2: Compute KL divergence for a review compared to the global average
# 리뷰의 토픽 비율과 전체 평균 토픽 비율 간의 KL-발산 계산
def compute_kl_divergence(review_text, model, vectorizer, global_topic_proportions):
    # 리뷰의 평균 토픽 비율 계산
    review_topic_proportions = aggregate_to_review_level(review_text, model, vectorizer)

    # KL-발산 계산
    kl_divergence = entropy(review_topic_proportions, global_topic_proportions)
    return kl_divergence

# Step 3: Apply Content Breadth calculation to each review
# DataFrame에 ContentBreadth 적용
global_topic_proportions = calculate_average_topic_proportions(df, lda_model, vectorizer)
df['ContentBreadth'] = df['Review_Text'].apply(
    lambda x: compute_kl_divergence(x, lda_model, vectorizer, global_topic_proportions)
)

# 결과 보기
df
```
